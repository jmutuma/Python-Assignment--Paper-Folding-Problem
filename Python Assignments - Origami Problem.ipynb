GRAPH VISUALIZATION
1. We create a list called thickness_list, add the value before folding(num) and append the values of folding paper in each step the confirm the no of values using len function.

```python
THICKNESS = 0.00008
num = THICKNESS
thickness_list = [num]
for n in range(43):
    num = num*2
    thickness_list.append(num)
print("Thickness: {: .2f} meters".format(num)) 
print("Thickness_list:",thickness_list)
print("Number of values in Thickness_list:",len(thickness_list))
    
```

2.Using number of folds as x values and thickness_list as y values we draw a  graph.

```python
import matplotlib.pyplot as plt # library used to draw the graph
%matplotlib inline
fig, ax = plt.subplots()
number_of_folds = [x for x in range(44)] # create a list of number of folds
ax.plot(number_of_folds, thickness_list)
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(thickness_list)
plt.show()
```

3. Customizing graphs:color line = green, increase line thickness, change font size of axis values

```python
import matplotlib.pyplot as plt # library used to draw the graph
%matplotlib inline
fig, ax = plt.subplots()
number_of_folds = [x for x in range(44)] # create a list of number of folds
ax.plot(number_of_folds, thickness_list)
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.tick_params(labelsize=20)
plt.plot(thickness_list, color = 'green', linewidth = 10)
plt.show()
```
